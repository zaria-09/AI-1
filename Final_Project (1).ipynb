```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
Gen1 = pd.read_csv("Gen1.csv")
Gen9 = pd.read_csv("Gen9.csv")
```

## üìä My Analysis

```python
print("5-number summary")
Gen1["HP"].describe()
```

```python
print("5-number summary")
Gen9["HP"].describe()
```

```python
sns.histplot(data=Gen9, x="HP", bins=15, color='skyblue', label='Generation 9', alpha=0.6)
sns.histplot(data=Gen1, x="HP", bins=15, label='Generation 1', alpha=0.6)
plt.title("Comparison of HP: Gen 1 vs. Gen 9", fontsize=16)
plt.xlabel("HP Value", fontsize=12)
plt.ylabel("Frequency (Number of Pokemon)", fontsize=12)
plt.legend() 
```

```python
Gen1["Generation"] = "Gen1"
Gen9["Generation"] = "Gen9"
Gen1_Gen9_combined = pd.concat([Gen1, Gen9], ignore_index=True)
custom_palette = {"Gen1": "tab:blue", "Gen9": "skyblue"}
sns.boxplot(data=Gen1_Gen9_combined, x='Generation', y='HP', hue="Generation",palette=custom_palette)
plt.title("Both generations", fontsize=15)
```

```python
# from five number summary 
average_gen1_hp = Gen1["HP"].mean()
average_gen1_hp
```

```python
average_gen9_hp = Gen9["HP"].mean()
average_gen9_hp 
```
