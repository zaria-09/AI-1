# What is the relationship between reviews that 300 S Broadway Denver, CO 80209 McDonald's have with Aurora, CO 80015Taco 16776 East Smokey Hill Road Taco Bell starting in the year of 2006 through year 2025?

The study I will be doing will include grouped data because I want all of the reviews that that specific McDonald's and that specific Taco Bells get within the time period of 2006 and 2025. The study will be doing is retrospective because the data I will be using is past data. My study will also be observational because I will be observing the data that I collect. The data I collected for both the specific McDonald's and Taco Bell was from Yelp, but yelp does not display reviews that it finds to be unhelpful, unreliable, or biased but these reviews can still be views but they aren't recommended. My data could be biased because I also used chatGPT to organize my data and it might have missed some of the data, but I did check it every time, though I could have missed something. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
McDonalds = pd.read_csv("McDonalds.csv")
TacoBell = pd.read_csv("TacoBell.csv")
```

```python
McDonalds["Year published"]
```

```python
McDonalds["Resturant"] = "McDonalds"
TacoBell["Resturant"] = "TacoBell"
Both_combined = pd.concat([McDonalds, TacoBell], ignore_index=True)
Both_combined['Year published'] = Both_combined['Year published'].astype(int)
sns.scatterplot(data=Both_combined, x='Year published', y='Name', hue="Resturant" )
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```

```python
sns.scatterplot(data=McDonalds, x="Year published", y= "Name")
```

```python
McDonalds_reviews = McDonalds["Year published"]
TacoBell_reviews = TacoBell["Year published"]
Both_reviews= pd.DataFrame({
    "McDonalds_Year_published" : McDonalds_reviews.reset_index(drop=True),
    "TacoBell_Year_published" : TacoBell_reviews.reset_index(drop=True)
})
sns.regplot(data=Both_reviews, x="McDonalds_Year_published", y="TacoBell_Year_published")
```

```python
McDonalds_reviews = McDonalds["Year published"]
TacoBell_reviews = TacoBell["Year published"]
```

```python
r = McDonalds_reviews.corr(TacoBell_reviews, method = 'pearson')
```

```python
display(r)
```

```python
import cmath
```

```python
r2 = cmath.sqrt(r)
```

```python
print(r2)
```

The relationship between reviews from that specific McDonald's and that specific Taco Bell is a negative linear relationship that is very very weak because the Pearson correlation coefficient (r) is around -0.004, and is very close to zero so there is almost no relationship between McDonald's reviews and Taco Bell reviews starting from 2006 to 2025.
