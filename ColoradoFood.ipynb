# What is the relationship between star ratings that 300 S Broadway Denver, CO 80209 McDonald's have with Aurora, CO 80015Taco 16776 East Smokey Hill Road Taco Bell starting in the year of 2006 through year 2025?

The study I will be doing will include a sample of data because I want all of the star rating of reviews that that specific McDonald's and that specific Taco Bells get within the time period of 2006 and 2025. The study will be doing is retrospective because the data I will be using is past data. My study will also be observational because I will be observing the data that I collect. The data I collected for both the specific McDonald's and Taco Bell was from Yelp, but yelp does not display reviews that it finds to be unhelpful, unreliable, or biased but these reviews can still be views but they aren't recommended. My data could be biased because I also used chatGPT to organize my data and it might have missed some of the data, but I did check it every time, though I could have missed something. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
McDonalds = pd.read_csv("McDonalds.csv")
TacoBell = pd.read_csv("TacoBell.csv")
```

```python
McDonalds["Name count"]
```

```python
ax = sns.regplot(data=TacoBell, x="Star rating", y= "count")
ax.set_ylabel("Name count of people who submited reviews")
```

```python
Comparison_data = pd.DataFrame({
    "McDonalds Star rating": McDonalds["Star rating"],
    "TacoBell Star rating": TacoBell["Star rating"]
})
ax = sns.regplot(data=Comparison_data, x="McDonalds Star rating", y= "TacoBell Star rating")
```

```python
McDonalds_star = McDonalds["Star rating"]
TacoBell_star = TacoBell["Star rating"]
```

```python
r = McDonalds_star.corr(TacoBell_star, method = 'pearson')
```

```python
display(r)
```

```python
import cmath
```

```python
r2 = cmath.sqrt(r)
```

```python
print(r2)
```

The relationship between Star ratings for that specific McDonald's and that specific Taco Bell is a positive linear relationship that is moderate to strong because the Pearson correlation coefficient (r) is around 0.692, so the relationship between McDonald's reviews and Taco Bell reviews starting from 2006 to 2025 is a positive and somewhat strong relationship.
