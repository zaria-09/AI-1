# What is the most common genre of music in Spotify U.S. playlists compared to the common genre of music in Spotify global playlists?

the study I will conduct will be observational study because I will be looking at data to compare the common genre of spotify U.S. playlists people listen to to the common genre of spotify global playlists. I will be using retrospective study. .the data was collected with Ai biases- some biases could be that ai isn't that accurate so some data might not be right.

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
us = pd.read_csv("US.csv")
world = pd.read_csv("Global.csv") #pie chart under
```

```python
data = {'AI Genre': ['Pop', 'Country', 'Hip-Hop', 'R&B','Indie Folk','Alternative Rock','Pop Rock','Indie Rock','Alternative Pop'],
        'RelativeFrequency as %': [16/50*100,11/50*100,9/50*100,5/50*100,3/50*100,3/50*100,1/50*100,1/50*100,1/50*100]}
df = pd.DataFrame(data)
```

```python
sns.set_style("dark")
plt.figure(figsize=(6,6))
plt.pie(df['RelativeFrequency as %'], labels=df['AI Genre'], autopct='%1.1f%%')
plt.title('Relative Frequency of U.S. playlists')
plt.show()
```

```python
data2 = {'AI Genre': ['Pop', 'Country', 'Hip-Hop', 'R&B','Indie Folk','Alternative Rock','Pop Rock','Indie Rock','Alternative','Latin','Indie Pop','K-Pop','Afrobeats'],
        'RelativeFrequency as %': [18/50*100,4/50*100,6/50*100,4/50*100,2/50*100,1/50*100,1/50*100,1/50*100,5/50*100,5/50*100,1/50*100,1/50*100,1/50*100]}
df = pd.DataFrame(data2)
```

```python
sns.set_style("dark")
plt.figure(figsize=(6,6))
plt.pie(df['RelativeFrequency as %'], labels=df['AI Genre'], autopct='%1.1f%%')
plt.title('Relative Frequency of Global playlists')
plt.show()
```

the most common genre of music from both playlists is Pop, with both pie charts showing Pop as the largest percentage then the rest. the percent of U.S. Pop genre being 32% and Global percent of Pop being 36%.
