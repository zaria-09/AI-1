# how do the average prices of Fiction books compare to the average prices of Non Fiction books?

The study I will be using is observational study because I will be observing the data. The data I will be using is sample data because the population of books is limited to only best seller books so it's only a sample. The data was collected by scraping Amazon’s best seller list. Any data that couldn’t be found in these lists was gathered using the GoodReads API. The lists were scraped in October 2020 to ensure that a decade of book data could be collected.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
books = pd.read_csv("books.csv")
```

```python
penguins = pd.read_csv("books.csv")
Fictionbooks = books[books["Genre"] == "Fiction"][["Price", "Genre"]]
sns.histplot(data=Fictionbooks, x="Price", bins=15)
plt.title("Prices of Amazon’s Bestselling Fiction Books", fontsize=15)
```

```python
penguins = pd.read_csv("books.csv")
NonFictionbooks = books[books["Genre"] == "Non Fiction"][["Price", "Genre"]]
sns.histplot(data=NonFictionbooks, x="Price", bins=15)
plt.title("Prices of Amazon’s Bestselling Non Fiction Books", fontsize=15)
```

```python
NonFictionbooks = books[books["Genre"] == "Non Fiction"][["Price", "Genre"]]
sns.boxplot(data=NonFictionbooks, y='Price',color='skyblue')
plt.title("Prices", fontsize=15)
```

```python
Fictionbooks = books[books["Genre"] == "Fiction"][["Price", "Genre"]]
sns.boxplot(data=Fictionbooks, y='Price')
plt.title("Prices", fontsize=15)
```

```python
Fictionbooks = books[books["Genre"] == "Fiction"][["Price", "Genre"]]
NonFictionbooks = books[books["Genre"] == "Non Fiction"][["Price", "Genre"]]
Both_Combined = pd.concat([Fictionbooks, NonFictionbooks], ignore_index=True) 
custom_palette = {"Fiction": "tab:blue", "Non Fiction": "skyblue"}
sns.boxplot(data=Both_Combined, x='Price', y='Genre',palette=custom_palette)
plt.title("Prices for Both Genres", fontsize=15)
```
